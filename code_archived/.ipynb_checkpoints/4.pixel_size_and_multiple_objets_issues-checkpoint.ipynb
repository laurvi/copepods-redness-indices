# How to integrate pixel size ? 

```python
# Read image and look at its size
imgs_calib_list = os.listdir("imgs_processed/test2/dataset1_calibrated") # for now, we are taking calibrated images with the first calibration method 
imgs_calib_list.sort()
filename_calibrated = "imgs_processed/test2/dataset1_calibrated/" + imgs_calib_list[9]
img_calib = io.imread(filename_calibrated)
plt.imshow(img_calib)

print('Type of the image : ' , type(img_calib)) 
print('Shape of the image : {}'.format(img_calib.shape)) 
print('Image Hight {}'.format(img_calib.shape[0])) 
print('Image Width {}'.format(img_calib.shape[1])) 
print('Dimension of Image {}'.format(img_calib.ndim))

print('Image size {}'.format(img_calib.size)) 


#display(np.mean(img_test, axis=(0)))
#display(np.mean(img_test, axis=(0,1))) # mean of distinct R, G and B intensities on the image > list of 3 values
#display(np.mean(img_test, axis=(0,1,2))) # display(np.mean(img_test)) : mean of total R, G and B intensisties over all pixels > only one 
```

```python
# Deconvolution matrix
rgb_from_stains = np.array([
    # RDG value from literature 
    [1.        ,  0.68396032,  0.6330574],
    # White
    [1, 1, 1],
    # Placeholder for third color
    [0,0,0]])
# Calculate residuals = normalize the OD matrix: 
rgb_from_stains[2, :] = np.cross(rgb_from_stains[0, :], rgb_from_stains[1, :])
display("Normalized OD matrix :", rgb_from_stains)

# Calculate matrix inverse with linalg.inv :
stains_from_rgb = linalg.inv(rgb_from_stains)
display("Inversed OD matrix = color deconvolution matrix:", stains_from_rgb)
```

```python
# Deconvolution by pixel
stains = separate_stains(img_calib, stains_from_rgb) # Extract stains 
red = -stains[...,0]# RGB to stain color space conversion (sickit image)
    
# Show the image with color deconvolution
plt.imshow(red)
plt.colorbar()
plt.show()
```

```python
# Normalise OD values between 0 and 1
img_norm = (red + np.absolute(np.min(red))) / (np.max(red) + np.absolute(np.min(red)))
plt.imshow(img_norm, vmin=0, vmax = 1)
plt.colorbar()
plt.show()
```

```python
# Threshold using intensities histogramm on the image 
thr = 0.40
red_pix = img_norm < thr
red_mask = img_norm > thr

# Distinguish red pixels on orginal image
img_highlight = copy.copy(img_calib)
img_highlight[:,:,0][red_mask] = 255
img_highlight[:,:,1][red_mask] = 255
img_highlight[:,:,2][red_mask] = 255

plt.imshow(img_highlight)
plt.show()
```

```python
# Look for red pixels on orginal image
img_reds = copy.copy(img_calib)
img_reds[:,:,0][red_pix] = 255
img_reds[:,:,1][red_pix] = 255
img_reds[:,:,2][red_pix] = 255
plt.imshow(img_reds)
plt.show()
```

```python
print('Image size {}'.format(img_reds.shape)) 
```

```python
print('There are', red_mask.sum(), 'astaxanthin red pixels, on a total of {}'.format(img_calib.shape[0]*img_calib.shape[1]), 'pixels')
```

```python
# Select only pixels with and OD value > 0.40
OD40plus = copy.copy(img_norm)
display(OD40plus)
display(red_mask)
display(OD40plus[red_mask])
```

```python
# Information of all astaxanthin pixels
print('number of red pixels :', OD40plus[red_mask].size )
print('total value of redness level :', OD40plus[red_mask].sum())
print('mean value of redness level :', OD40plus[red_mask].mean())
```

```python
# Information on all non-astaxanthin pixels
print('number of red pixels :', OD40plus[red_pix].size )
print('total value of redness level :', OD40plus[red_pix].sum())
print('mean value of redness level :', OD40plus[red_pix].mean())
```
